```python
#Step 5: Handle Missing Values
#Handle price column:
#Reason: price is an essential variable for any analysis involving property listings. Missing prices can lead to inaccurate results.
#Approach: Impute missing values with the median price. Median is less sensitive to extreme outliers than the mean.
#Imputing missing values ensures that I donâ€™t lose valuable data from the dataset while maintaining a reasonable estimate of what the price would be.
```

```python
listings_data['price'] = listings_data['price'].fillna(listings_data['price'].median()) 
```

```python
# Handling reviews_per_month column:
# Reason: If a listing has no reviews, it may mean the property is new or hasnâ€™t been booked much.
# Approach:
# Impute missing values with 0 because itâ€™s logical that properties with no reviews should be treated as having no reviews per month.
# Justification: This approach prevents you from losing rows of data due to missing values, while accounting for the properties that havenâ€™t had reviews yet.
```

```python
listings_data['reviews_per_month'] = listings_data['reviews_per_month'].fillna(0)
```

```python
# Step 9: Detecting and Handling Outliers Using Z Score Method
# We can deal with outliers in various ways. 
# One method is to calculate the z-score (how many standard deviations above or below the mean a data point is) 
# and drop any observation that has a z-score above 2.53 or below -2.53. 
# Note that whether or not to remove outliers should be determined based on the business context.
```

```python
# Import the necessary library for Z-score calculation
from scipy import stats
from scipy.stats import zscore
import numpy as np
# Step: Detect and Handle Outliers Using Z-score
from scipy.stats import zscore

# Step 1: Calculate Z-scores for 'price' column
listings_data['z-score_price'] = zscore(listings_data['price'])
print("Price Z-score stats:")
print("Max Z-score (Price):", listings_data['z-score_price'].max())
print("Min Z-score (Price):", listings_data['z-score_price'].min())

# Step 2: Calculate Z-scores for 'minimum_nights' column
listings_data['z-score_minimum_nights'] = zscore(listings_data['minimum_nights'])
print("\nMinimum Nights Z-score stats:")
print("Max Z-score (Minimum Nights):", listings_data['z-score_minimum_nights'].max())
print("Min Z-score (Minimum Nights):", listings_data['z-score_minimum_nights'].min())

# Step 3: Calculate Z-scores for 'number_of_reviews' column
listings_data['z-score_number_of_reviews'] = zscore(listings_data['number_of_reviews'])
print("\nNumber of Reviews Z-score stats:")
print("Max Z-score (Number of Reviews):", listings_data['z-score_number_of_reviews'].max())
print("Min Z-score (Number of Reviews):", listings_data['z-score_number_of_reviews'].min())

# Step 4: Calculate Z-scores for 'reviews_per_month' column
listings_data['z-score_reviews_per_month'] = zscore(listings_data['reviews_per_month'])
print("\nReviews per Month Z-score stats:")
print("Max Z-score (Reviews per Month):", listings_data['z-score_reviews_per_month'].max())
print("Min Z-score (Reviews per Month):", listings_data['z-score_reviews_per_month'].min())

# Step 5: Calculate Z-scores for 'availability_365' column
listings_data['z-score_availability_365'] = zscore(listings_data['availability_365'])
print("\nAvailability 365 Z-score stats:")
print("Max Z-score (Availability 365):", listings_data['z-score_availability_365'].max())
print("Min Z-score (Availability 365):", listings_data['z-score_availability_365'].min())
```

```python
# Define the threshold for outliers
threshold = 3

# Identify outliers in each column based on Z-scores
outliers_price = listings_data[listings_data['z-score_price'].abs() > threshold]
outliers_minimum_nights = listings_data[listings_data['z-score_minimum_nights'].abs() > threshold]
outliers_number_of_reviews = listings_data[listings_data['z-score_number_of_reviews'].abs() > threshold]
outliers_reviews_per_month = listings_data[listings_data['z-score_reviews_per_month'].abs() > threshold]
outliers_availability_365 = listings_data[listings_data['z-score_availability_365'].abs() > threshold]

# Print outliers found
print(f"Price outliers:\n{outliers_price}")
print(f"Minimum Nights outliers:\n{outliers_minimum_nights}")
print(f"Number of Reviews outliers:\n{outliers_number_of_reviews}")
print(f"Reviews per Month outliers:\n{outliers_reviews_per_month}")
print(f"Availability 365 outliers:\n{outliers_availability_365}")
```

```python
# Defining the threshold for outliers
threshold = 2.53

# Apply the filter for each Z-score column
listings_data = listings_data[                                                #removing all rows with a z-score above 2.53 or below -2.53
    (abs(listings_data["z-score_price"]) <= threshold) &
    (abs(listings_data["z-score_minimum_nights"]) <= threshold) &
    (abs(listings_data["z-score_number_of_reviews"]) <= threshold) &
    (abs(listings_data["z-score_reviews_per_month"]) <= threshold) &
    (abs(listings_data["z-score_availability_365"]) <= threshold)]

# Checking the shape of the cleaned dataset
print(listings_data.shape)
```
